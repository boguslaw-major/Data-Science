```
import numpy as np
import matplotlib.pyplot as plt
x=range(-10,10)
y=[]
for i in x:
  y.append(2*i-3)



#x=np.arange(-3,3,0.1)
#y=np.sin(x)

plt.plot(x,y)
plt.title("Wykres y=sinx")
plt.show()
```
